```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
x=I = np.linspace(10,90,9) 
y=V = np.array([0.98,1.98,2.98, 3.97,4.95,5.95, 6.93,7.93,8.91])
```

```python
N = len(I)
delta = N *np.sum(x*x)-np.sum(x)**2
c = (np.sum(x*x)*np.sum(y)-np.sum(x)*np.sum(x*y))/delta
m = (N*np.sum(x*y)-np.sum(x)*np.sum(y))/delta
alpha_cu = np.sqrt((1/(N-2))*np.sum((y-m*x-c)**2))
alpha_c = alpha_cu *np.sqrt(np.sum(x*x)/delta)
alpha_m = alpha_cu*np.sqrt(N/delta)
print(c)
print(m)
print(alpha_cu)
print(alpha_c)
print(alpha_m)
```

```python
V_ajustado = m * I + c

# Calcular residuos
residuos = V - V_ajustado
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))

# Ajusta el espacio entre las gráficas
fig.subplots_adjust(hspace=0.1)  # Ajusta este valor según necesites

# Gráfica 1
ax1.errorbar(I, V, yerr=0.01, fmt='o', color = "black", label = "Valores")
ax1.plot(I,V_ajustado, )
ax1.set_ylabel('V [mV]')
ajuste_texto = r'$y = (0.09908 \pm 9e^{-5})x - (1 \pm 5)e^{-3}  $'
ax1.text(0.05, 0.95, ajuste_texto, transform=ax1.transAxes, fontsize=12, verticalalignment='top')


# Gráfica 2
ax2.scatter(I,residuos)
ax2.set_xlabel('I [μA]')
ax2.set_ylabel('Residuos')
```
