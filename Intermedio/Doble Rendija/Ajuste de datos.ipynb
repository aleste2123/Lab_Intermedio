Ajuste una rendija

```python
#campana de gauss equación
def gauss(x,a,b,c):
    return a*np.exp(-((x-b)**2)/(2*c**2))
res_1, con_1 = curve_fit(gauss,Rotacion_una_rendija,Intensidad_una_rendija,p0=[np.max(Intensidad_una_rendija),np.max(Rotacion_una_rendija)/2,1000])
guass_fit = gauss(Rotacion_una_rendija,*res_1)
```

Ajuste de dos rendijas

```python
def ajuste(theta,I,lamb,d,a,b,c,e):
    f =I*np.cos(np.sin((theta-c)/b)*np.pi*d/lamb)**2
    g =I*(np.sin(np.pi*a*np.sin((theta-c)/b)/lamb)/(np.pi*a*np.sin((theta-c)/b)/lamb))**2
    return f*g+e




res_2, con_2 = curve_fit(ajuste,Rotacion_filtro,filtro_doble_rendija,p0=[0.4,680*10**(-9),50*10**(-6),2.6e-06,999,1,0.01])

#funciona[0.4,680*10**(-9),50*10**(-6),2.6e-06,999,1,0.01]
#res_3, con_3 = curve_fit(sinc_cos,Rotacion_filtro,filtro_doble_rendija,p0=[np.max(filtro_doble_rendija)**(1/2),60,0,10])

doble_rendija_fit = ajuste(Rotacion_filtro,*res_2)
print(np.sum(residuos(filtro_doble_rendija,doble_rendija_fit)))
print(res_2)
#doble_rendija_fit_2 = sinc_cos(Rotacion_filtro,*res_3)
plt.plot(Rotacion_filtro,doble_rendija_fit)
#plt.plot(Rotacion_filtro,doble_rendija_fit_2)
```
